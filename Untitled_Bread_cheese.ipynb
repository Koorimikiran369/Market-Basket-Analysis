```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
import mlxtend as ml
from mlxtend.preprocessing import TransactionEncoder
import time
```

```python
f = open("Bread_Cheese.csv")
filedata = f.readlines()

Items = []

for x in filedata:
    x = x.strip('\n').split(',')
    Items.append(x)

print(Items)
```

```python
df = pd.DataFrame(Items)
df
```

```python
te = TransactionEncoder()
te_array = te.fit(Items).transform(Items)
pd.DataFrame(te_array.head()
```

```python
te.columns_mapping_
```

```python
te.columns_
```

```python
df = pd.DataFrame(te_array.astype(int), columns = te.columns_)
df
```

```python
# Generate frequent item sets
x = time.time()
freq_itemsets = apriori(df, min_support=0.16,use_colnames=True)
# Use min support to start with
y = time.time()
print('Time in seconds %s'%(y-x))
```

```python
freq_itemsets
```

```python
rules = association_rules(freq_itemsets)
rules.shape
```

```python
rules[(rules['lift']>=2)].iloc[:,0:7]
```
